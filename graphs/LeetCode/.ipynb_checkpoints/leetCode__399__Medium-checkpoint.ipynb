```python
class Solution:
    def calcEquation(self, equations: List[List[str]], values: List[float], queries: List[List[str]]) -> List[float]:
        
        graph = defaultdict(defaultdict)
        
        for (dividend,divisor), value in zip(equations,values):
            graph[dividend][divisor] = value
            graph[divisor][dividend] = 1 / value
        
        print(graph)
        
        
        myGraph = {}
        
        for (dividend,divisor), value in zip(equations,values):
            
            if myGraph.get(dividend,-1) == -1 :
                myGraph[dividend] = {divisor : value}
            else:
                myGraph[dividend][divisor] = value
                
            if myGraph.get(divisor,-1) == -1 :
                myGraph[divisor] = {dividend : 1/value}
            else:
                myGraph[divisor][dividend] = 1/value
                
                
        # print(myGraph)
        
        def f(curr_node, target_node, acc_product, visited):
            
            visited.add(curr_node)
            
            ret = -1.0
            
            neighbours = myGraph[curr_node]
            
            if target_node in neighbours:
                ret = acc_product * neighbours[target_node]
            
            else:
                
                for neighbour,value in neighbours.items():
                    
                    if neighbour in visited :
                        continue
                    ret = f(neighbour, target_node, acc_product * value, visited)
                    
                    if ret != -1.0:
                        break
            
            visited.remove(curr_node)
            
            return ret
        
        results = []
        
        for dividend,divisor in queries:
            
            if dividend not in myGraph or divisor not in myGraph:
                ret = -1.0
            elif dividend == divisor :
                ret = 1.0
            else:
                visited = set()
                ret = f(dividend,divisor,1, visited)
            
            results.append(ret)
            
#         print(results)
        
        return results
        
```
