```python
from collections import defaultdict
import heapq
```

```python
def minCostToSupplyWater( n , wells, pipes):

    # Step 1 : Building the graph 

    graph = defaultdict(list)

    for index,cost in enumerate(wells):
        graph[0].append((cost,index+1))

    for house_1, house_2,cost in pipes :
        graph[house_1].append((cost,house_2))
        graph[house_2].append((cost,house_1))

    print("\nGraph : ", graph)

    # Step 2 : Maintain a set to track visited list of vertices

    mst_set = set([0])
    print("\nmst_set : ", mst_set)

    # Step 3 : 

    heapq.heapify(graph[0])
    print("\nGraph : ", graph)

    edges_heap = graph[0]
    print(edges_heap)

    total_cost = 0 

    while len(mst_set) < (n+1):
        break



    return -1
```

```python
n = 3
wells = [1,12,7]
pipes = [[1,2,1],[2,3,1]]
```

```python
minCostToSupplyWater( n , wells, pipes)
```
